Voy a calcular el porcentaje de NaN por columna por si tengo que modificarlas.

```python
def calcular_nan(df):
  total_filas = len(df)
  nan_por_columna = df.isnull().sum()
  porcentaje_nan = (nan_por_columna / total_filas) * 100
  return porcentaje_nan

df_nan = calcular_nan(df_copy)
print(df_nan)
```
